```python
import pandas as pd

housing_df = pd.read_csv(r"C:\Users\vasup\Downloads\housing_data.csv")
transactions_df = pd.read_csv(r"C:\Users\vasup\Downloads\transactions_data.csv")

housing_df.head(), transactions_df.head()
```

```python
housing_df.describe()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(housing_df["price"],kde=True)
plt.title("Distribution of House prices")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x="size_sqft",
    y="price",
    data=housing_df
)
plt.title("House size vs price")
plt.xlabel("Size (sq ft)")
plt.ylabel("Price")
plt.show()
    
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    x="rooms",
    y="price",
    data=housing_df)
plt.title("Number of rooms vs price")
plt.xlabel("Number of rooms")
plt.ylabel("price")
plt.show()
```

```python
housing_df.corr()
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(
    housing_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Matrix - Housing Data")
plt.show()
```
